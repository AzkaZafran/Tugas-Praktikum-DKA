### Jawaban Latihan Soal
#### Pre-Process:

```python
import matplotlib.pyplot as plt
import networkx as nx
```

```python
def show_graph(G, pos=None, title='') :
    if pos == None :
        pos = nx.spring_layout(G)

    plt.figure(figsize=(15, 15))
    nx.draw(
        G,
        pos,
        with_labels=True,
        node_color="red",
        node_size=1500,
        font_color="black",
        font_weight="bold",
        width=5
    )

    edge_labels = nx.get_edge_attributes(G, "weight")

    nx.draw_networkx_edge_labels(
        G,
        pos,
        edge_labels=edge_labels,
        font_color="blue",
        font_weight="bold",
        font_size=12
    )

    plt.margins(0.2)
    plt.title(title)
    plt.show()
```

#### Jawaban Soal 1a:

```python
posEropa = {
    'Arad': (0, 10),
    'Timisonara': (-6, 8),
    'Sibiu': (0, 8),
    'Zerind': (6, 8),
    'Lugoj': (-6, 6),
    'Rimnicu Vilsea': (-2, 6),
    'Fagaras': (2, 6),
    'Oradea': (6, 6),
    'Mehadia': (-6, 4),
    'Craiova': (-3, 4),
    'Pitesti': (-1, 4),
    'Bucharest': (2, 4),
    'Drobeta': (-6, 2),
    'Glurgiu': (1, 2),
    'Urziceni': (3, 2),
    'Hirsova': (2, 0),
    'Vaslui': (4, 0),
    'Eforie': (2, -2),
    'Iasi': (4, -2),
    'Neamt': (4, -4)
}

nodesEropa_data = [x for x in posEropa.keys()]

edgesEropa_weighted_data = [('Arad', 'Timisonara', 118), ('Arad', 'Sibiu', 140), ('Arad', 'Zerind', 75),
                   ('Timisonara', 'Lugoj', 111), ('Sibiu', 'Rimnicu Vilsea', 80), ('Sibiu', 'Fagaras', 99),
                   ('Sibiu', 'Oradea', 151), ('Zerind', 'Oradea', 71), ('Lugoj', 'Mehadia', 70),
                   ('Rimnicu Vilsea', 'Craiova', 146), ('Rimnicu Vilsea', 'Pitesti', 97), ('Fagaras', 'Bucharest', 211),
                   ('Mehadia', 'Drobeta', 75), ('Craiova', 'Pitesti', 138), ('Pitesti', 'Bucharest', 101), ('Bucharest', 'Glurgiu', 90), ('Bucharest', 'Urziceni', 85),
                   ('Drobeta', 'Craiova', 120), ('Urziceni', 'Hirsova', 98), ('Urziceni', 'Vaslui', 142),
                   ('Hirsova', 'Eforie', 86), ('Vaslui', 'Iasi', 92), ('Iasi', 'Neamt', 87)]

GEropa = nx.Graph()
GEropa.add_nodes_from(nodesEropa_data)
GEropa.add_weighted_edges_from(edgesEropa_weighted_data)

show_graph(GEropa, posEropa, title='Graf Eropa')

print("Urutan node yang dikunjungi: ")
i = 1
for nodes in nx.bfs_edges(GEropa, source='Arad'):
    print(f'{i}. {nodes[0]} - {nodes[1]}')
    i += 1
```

#### Jawaban Soal 1b:

```python
posIndonesia = {
    'Bandung': (0, 10),
    'Yogyakarta': (-6, 8),
    'Cirebon': (0, 8),
    'Jakarta': (6, 8),
    'Surakarta': (-8, 6),
    'Semarang': (-4, 6),
    'Malang': (-8, 4),
    'Surabaya': (-4, 4)
}

nodesIndonesia_data = [x for x in posIndonesia.keys()]

edgesIndonesia_Weighted_data = [('Bandung', 'Yogyakarta', 373), ('Bandung', 'Cirebon', 120), ('Bandung', 'Jakarta', 270),
                       ('Yogyakarta', 'Surakarta', 60), ('Yogyakarta', 'Semarang', 109), ('Yogyakarta', 'Cirebon', 210),
                       ('Cirebon', 'Jakarta', 327), ('Surakarta', 'Malang', 370), ('Surakarta', 'Semarang', 97),
                       ('Semarang', 'Surabaya', 369), ('Malang', 'Surabaya', 94)]

GIndonesia = nx.Graph()
GIndonesia.add_nodes_from(nodesIndonesia_data)
GIndonesia.add_weighted_edges_from(edgesIndonesia_Weighted_data)

show_graph(GIndonesia, posIndonesia, title='Graf Indonesia')

print("Urutan node yang dikunjungi: ")
i = 1
for nodes in nx.bfs_edges(GIndonesia, source='Bandung'):
    print(f'{i}. {nodes[0]} - {nodes[1]}')
    i += 1
```
